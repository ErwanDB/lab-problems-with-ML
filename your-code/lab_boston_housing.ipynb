## Developing a Model

### Implementation: Define a Performance Metric
What is the performance metric with which you will determine the performance of your model? Create a function that calculates this performance metric, and then returns the score. 

```python
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    # Your code here:
    x = r2_score(y_true, y_predict)
    return x
```

### Implementation: Shuffle and Split Data
Split the data into the testing and training datasets. Shuffle the data as well to remove any bias in selecting the train and test. 

```python
from sklearn.model_selection import train_test_split

train,test=train_test_split(houses,train_size=0.8,shuffle=True)

X_train = train.drop(["medv"],axis=1)
X_test = test.drop(["medv"],axis=1)

y_train = train["medv"]
y_test = test["medv"]
```

----

## Analyzing Model Performance
Next, we are going to build a Random Forest Regressor, and test its performance with several different parameter settings.

### Learning Curves
Lets build the different models. Set the max_depth parameter to 2, 4, 6, 8, and 10 respectively. 

```python
from sklearn.ensemble import RandomForestRegressor

# Five separate RFR here with the given max depths

RF_2 = RandomForestRegressor(max_depth=2)
RF_2.fit(X_train,y_train)

RF_4 = RandomForestRegressor(max_depth=4)
RF_4.fit(X_train,y_train)

RF_6 = RandomForestRegressor(max_depth=6)
RF_6.fit(X_train,y_train)

RF_8 = RandomForestRegressor(max_depth=8)
RF_8.fit(X_train,y_train)

RF_10 = RandomForestRegressor(max_depth=10)
RF_10.fit(X_train,y_train)
```

Now, plot the score for each tree on the training set and on the testing set.

```python
# Produce a plot with the score for the testing and training for the different max depths
score_train = []
score_test = []

for i in range(2,12,2):
    model = RandomForestRegressor(max_depth=i)
    model.fit(X_train,y_train)
    score_train.append(r2_score(y_train,model.predict(X_train)))
    score_test.append(r2_score(y_test,model.predict(X_test)))

#plt.figure(figsize=(10,8))
plt.plot(range(2,12,2),score_train)
plt.plot(range(2,12,2),score_test)
plt.xlabel('max depth random forest')
plt.ylabel('r2 score')
plt.legend(['train','test'])
plt.show()
```
